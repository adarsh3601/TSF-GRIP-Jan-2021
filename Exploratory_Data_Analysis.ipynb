```python
plt.figure(figsize=(10,10))
df['Sub-Category'].value_counts().plot.pie()
plt.show()
```

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(12,12))
sns.countplot(x= 'State', data=df, palette='rocket_r',order=df['State'].value_counts().index)
plt.xticks(rotation= 90)
plt.show()
```

```python
df.hist(figsize= (20,10), bins=50)
plt.show()
```

```python
plt.figure(figsize=(10,10))
df['Region'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
style.use('dark_background')
fig, ax =plt.subplots(figsize=(12,8))
ax.scatter(df['Sales'],df['Profit'])
ax.set_xlabel('Scales')
ax.set_ylabel('Profit')
plt.show()
```

```python
sns.lineplot(x= 'Discount', y= 'Profit', label='Profit', data=df)
plt.legend()
plt.show()
```

```python
#Profit vs Quantity
sns.lineplot(x= 'Quantity', y = 'Profit', label='Profit', data=df)
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['red','blue'], figsize=(10,10))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette= 'rocket')
plt.xlabel('Region', fontsize= 15)
plt.show()
```

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['yellow','green'], figsize=(10,10))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
ps = df.groupby('State')[['Profit','Sales']].sum().sort_values(by='Sales', ascending= False)
ps[:].plot.bar(color=['pink','blue'],figsize=(15,8))
plt.title('Profit/Loss and Sales in States')
plt.xlabel('States')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
top_states = df['State'].value_counts().nlargest()
top_states
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['yellow','green'],alpha=0.9,figsize=(10,10))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].mean().sort_values(by= 'Sales',ascending= False)
ps[:].plot.bar(color=['red','purple'],figsize=(10,10))
plt.title('Profit/Loss and Sales in states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```
